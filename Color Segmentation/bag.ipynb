```python
import matplotlib.pyplot as plt
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
from matplotlib import colors
import cv2

img =cv2.imread('bags.png')
plt.imshow(img)
plt.show()
img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
r, g, b= cv2.split(img)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")
pixel_colors = img.reshape((np.shape(img)[0]*np.shape(img)[1],3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
axis.scatter(r.flatten(),g.flatten(),b.flatten(),facecolors=pixel_colors,marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
light_yellow = (255, 254, 30)
dark_yellow = (166, 130, 1)

from matplotlib.colors import hsv_to_rgb
lo_square = np.full((10, 10, 3), light_yellow, dtype=np.uint8) /255.0
do_square = np.full((10, 10, 3), dark_yellow, dtype=np.uint8) /255.0
plt.subplot(1, 2, 1)
plt.imshow(lo_square)
plt.subplot(1, 2, 2)
plt.imshow(do_square)
plt.show()
```

```python
mask = cv2.inRange(img,dark_yellow,light_yellow)
y1  = cv2.bitwise_and(img,img,mask=mask)
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(y1)
plt.show()
```

```python
from skimage.io import imread,imshow 
from skimage.color import rgb2hsv

hsv_img = rgb2hsv(img)
fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(hsv_img[:,:,0],cmap='hsv')
ax[0].set_title('hue')
ax[1].imshow(hsv_img[:,:,1],cmap='hsv')
ax[1].set_title('transparecy')
ax[2].imshow(hsv_img[:,:,2],cmap='hsv')
ax[2].set_title('value')
fig.colorbar(imshow(hsv_img[:,:,2],cmap='hsv'))
#fig.tight_layout()
plt.show()
```

```python
hsv_img = rgb2hsv(img)
#refer to hue channel (in the colorbar)
lower_mask = hsv_img[:,:,0]>0.25
#refer to hue channel (in the colorbar)
upper_mask = hsv_img[:,:,0]<0.5
#refer to transparency (in the colorbar)
saturation_mask = hsv_img [:,:,1]>0.1

mask2 = upper_mask*lower_mask*saturation_mask
red = img[:,:,0]*mask2
green = img[:,:,1]*mask2
blue = img[:,:,2]*mask2
g = np.dstack((red,green,blue))
imshow(g)
```
